```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load data into a DataFrame
data = {
    'ID': [5524, 2174, 4141, 6182, 5324, 7446, 965, 6177, 4855, 5899, 1994, 387, 2125, 8180, 2569, 2114, 9736, 4939, 6565, 2278],
    'Year_Birth': [1957, 1954, 1965, 1984, 1981, 1967, 1971, 1985, 1974, 1950, 1983, 1976, 1959, 1952, 1987, 1946, 1980, 1946, 1949, 1985],
    'Income': [58138, 46344, 71613, 26646, 58293, 62513, 55635, 33454, 30351, 5648, np.nan, 7500, 63033, 59354, 17323, 82800, 41850, 37760, 76995, 33812],
    'MntWines': [635, 11, 426, 11, 173, 520, 235, 76, 14, 28, 5, 6, 194, 233, 3, 1006, 53, 84, 1012, 4],
    'MntFruits': [88, 1, 49, 4, 43, 42, 65, 10, 0, 0, 5, 16, 61, 2, 14, 22, 5, 5, 80, 17],
    'MntMeatProducts': [546, 6, 127, 20, 118, 98, 164, 56, 24, 6, 6, 11, 480, 53, 17, 115, 19, 38, 498, 19],
    'MntFishProducts': [172, 2, 111, 10, 46, 0, 50, 3, 3, 1, 0, 11, 225, 3, 6, 59, 2, 150, 0, 30],
    'MntSweetProducts': [88, 1, 21, 5, 27, 42, 49, 1, 3, 1, 2, 1, 112, 5, 1, 68, 13, 12, 16, 24],
    'MntGoldProds': [88, 6, 42, 5, 15, 14, 27, 23, 2, 13, 0, 16, 30, 14, 5, 45, 4, 28, 176, 39],
    'NumDealsPurchases': [3, 2, 1, 2, 5, 2, 4, 2, 1, 1, 1, 1, 1, 3, 1, 1, 3, 2, 2, 2],
    'NumWebPurchases': [8, 1, 8, 2, 5, 6, 7, 4, 0, 0, 2, 2, 3, 6, 1, 7, 3, 4, 11, 2],
    'NumCatalogPurchases': [10, 1, 2, 0, 3, 4, 3, 0, 0, 0, 0, 0, 4, 1, 0, 6, 0, 1, 4, 1],
    'NumStorePurchases': [4, 2, 10, 4, 6, 10, 7, 4, 2, 0, 2, 3, 8, 5, 3, 12, 3, 6, 9, 3],
    'NumWebVisitsMonth': [7, 5, 4, 6, 5, 6, 6, 8, 9, 20, 2, 8, 2, 6, 8, 12, 0, 7, 5, 6]
}

df = pd.DataFrame(data)

# Fill missing Income with median value for simplicity
df['Income'].fillna(df['Income'].median(), inplace=True)

# Normalize the relevant columns
features = ['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
df_features = df[features]

# Standardizing the data
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_features), columns=features)

# Calculate Jaccard Coefficient
def jaccard_coefficient(v1, v2):
    intersection = np.sum(np.minimum(v1, v2))
    union = np.sum(np.maximum(v1, v2))
    return intersection / union

# Calculate Simple Matching Coefficient
def simple_matching_coefficient(v1, v2):
    return np.sum(v1 == v2) / len(v1)

# Calculate Cosine Similarity
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    return dot_product / (norm_v1 * norm_v2)

# Compute pairwise similarities
num_vectors = len(df_scaled)
jc_matrix = np.zeros((num_vectors, num_vectors))
smc_matrix = np.zeros((num_vectors, num_vectors))
cos_matrix = np.zeros((num_vectors, num_vectors))

for i in range(num_vectors):
    for j in range(num_vectors):
        if i != j:
            jc_matrix[i, j] = jaccard_coefficient(df_scaled.iloc[i].values, df_scaled.iloc[j].values)
            smc_matrix[i, j] = simple_matching_coefficient(df_scaled.iloc[i].values, df_scaled.iloc[j].values)
            cos_matrix[i, j] = cosine_similarity(df_scaled.iloc[i].values, df_scaled.iloc[j].values)

# Create heatmaps for visualization
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

sns.heatmap(jc_matrix, ax=axs[0], cmap='viridis', annot=True, fmt='.2f')
axs[0].set_title('Jaccard Coefficient')

sns.heatmap(smc_matrix, ax=axs[1], cmap='viridis', annot=True, fmt='.2f')
axs[1].set_title('Simple Matching Coefficient')

sns.heatmap(cos_matrix, ax=axs[2], cmap='viridis', annot=True, fmt='.2f')
axs[2].set_title('Cosine Similarity')

plt.tight_layout()
plt.show()
```
